## Coding Session II - 30 minutes

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
#Esta parte me ha costado bastante sacarla...lo he hecho con ayuda de los compañeros

time=[fitbit.loc[:,"Minutes Lightly Active"], fitbit.loc[:,"Minutes Very Active"], fitbit.loc[:,"Minutes Sedentary"]]


for time in time:
    sns.scatterplot( x = time, y = fitbit.Steps)
    plt.show()
```
