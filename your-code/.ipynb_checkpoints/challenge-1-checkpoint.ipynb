**0. Load seaborn dataset "tips"**

```python
import pandas as pd 
import seaborn as sns
```

```python
tips= sns.load_dataset("tips")
```

```python
tips
```

**1. What is the prefered day for eating?**

```python
tips['day'].mode()
#el día más repetido para comer es el sábado
```

```python
sns.countplot(x=tips.day)
```

**2. What days of the week has more men than women payers?**

```python
sns.countplot(x=tips.day,
              hue=tips.sex,
              palette="Blues")
#los viernes, sábados y domingos son los días en los cuales hay más hombres que mujeres pagando
```

**3. Do you think bill amounts change considerably depending on smoking?** #relacion entre si fuman y el coste de la factura

```python
sns.countplot(x=tips.total_bill,
              hue=tips.smoker,
              palette="Reds")
#Podemos observar, obviando desviaciones puntuales, que el hecho de ser fumador no implica cambios considerables en la cuenta
```

**4. Is tip correlated to total_bill? Draw a scatter plot**

```python
tips.corr()
#Si podríamos decir que hay una correlación entre el total de la cuenta y la propina que se deja (cuanto mas grande es la cuenta,mas propina se deja), aunque no llega a ser una correlación excesivamente fuerte.
#añado scatterplot y boxplot en donde se puede observar
```

```python
sns.scatterplot(y="tip",
                x="total_bill",
                data=tips)
```

```python
sns.boxplot(x=tips.total_bill,
            y=tips.tip,
            data=tips)
```
